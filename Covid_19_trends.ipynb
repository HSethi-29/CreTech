```python
# Data Cleaning
mis_values=covid_data.isnull().sum()   # Missing Values count
```

```python
mis_values=mis_values[mis_values>0]
print(mis_values.to_string())
```

```python
# Handling Missing Values
drop_cols=['excess_mortality','excess_mortality_cumulative','excess_mortality_cumulative_absolute','excess_mortality_cumulative_per_million',
           'weekly_icu_admissions','weekly_icu_admissions_per_million','weekly_hosp_admissions','weekly_hosp_admissions_per_million']
covid_data.drop(columns=drop_cols,inplace=True)
```

```python
covid_data.info()
```

```python
# Forward-fill core time series data

# Sort before filling
covid_data.sort_values(['location', 'date'], inplace=True)

# Forward-fill and backfill for smoother results
covid_data=covid_data.ffill()
covid_data=covid_data.bfill()
```

```python
covid_data.head()
```

```python
# Data Visualisation

# Group by date to get global totals
global_trend = covid_data.groupby('date')[['new_cases', 'new_deaths']].sum().reset_index()

# Convert 'date' to datetime if not already
global_trend['date'] = pd.to_datetime(global_trend['date'])
```

```python
plt.figure(figsize=(6,4))
# Plot New cases
sns.lineplot(x='date',y='new_cases',data=global_trend,label='New Cases')

# Plot New Deaths
sns.lineplot(x='date',y='new_deaths',data=global_trend,label='New Deaths')

# Titles and labels
plt.title('Global Daily New COVID-19 Cases and Deaths', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Group the data to get the latest total cases for each country
latest_data = covid_data.groupby('location').agg({'date': 'max'}).reset_index()
latest_data = latest_data.merge(covid_data, on=['location', 'date'], how='left')

# Remove entries like 'World', 'Africa', etc. (they are aggregates, not countries)
excluded = ['World', 'Africa', 'Asia', 'Europe', 'European Union', 
            'International', 'North America', 'Oceania', 'South America']
latest_data = latest_data[~latest_data['location'].isin(excluded)]

# Sort and get top 10 countries by total cases
top_10_cases = latest_data.sort_values('total_cases', ascending=False).head(10)

# Plotting
plt.figure(figsize=(6,4))
sns.barplot(data=top_10_cases, x='total_cases', y='location', hue='location', palette='viridis', dodge=False, legend=False)

plt.title('Top 10 Countries by Total COVID-19 Cases', fontsize=16)
plt.xlabel('Total Cases')
plt.ylabel('Country')
plt.tight_layout()
plt.show()
```

```python
# Group by 'location' to get the latest date per country
latest_deaths = covid_data.groupby('location').agg({'date': 'max'}).reset_index()

# Merge with original dataset to get all metrics for that latest date
latest_deaths = latest_deaths.merge(covid_data, on=['location', 'date'], how='left')

# Exclude regions and continents
excluded = ['World', 'Africa', 'Asia', 'Europe', 'European Union',
            'International', 'North America', 'Oceania', 'South America']
latest_deaths = latest_deaths[~latest_deaths['location'].isin(excluded)]

# Sort by total deaths and get the top 10
top_10_deaths = latest_deaths.sort_values('total_deaths', ascending=False).head(10)

# Plotting
plt.figure(figsize=(6,4))
sns.barplot(data=top_10_deaths, x='total_deaths', y='location',
            hue='location', palette='magma', dodge=False, legend=False)

plt.title('Top 10 Countries by Total COVID-19 Deaths', fontsize=16)
plt.xlabel('Total Deaths')
plt.ylabel('Country')
plt.tight_layout()
plt.show()
```

```python
#Correlation Between Cases and Deaths

#Prepare the most recent data for each country
latest_data = covid_data.groupby('location').agg({'date': 'max'}).reset_index()
latest_data = latest_data.merge(covid_data, on=['location', 'date'], how='left')

# Step 2: Filter out non-country entries
excluded = ['World', 'Africa', 'Asia', 'Europe', 'European Union',
            'International', 'North America', 'Oceania', 'South America']
latest_data = latest_data[~latest_data['location'].isin(excluded)]

# Drop rows with missing total_cases or total_deaths
latest_data = latest_data.dropna(subset=['total_cases', 'total_deaths'])

# Create scatter plot to visualize correlation
plt.figure(figsize=(6,4))
sns.scatterplot(data=latest_data, x='total_cases', y='total_deaths', hue='continent', palette='Set1', alpha=0.7)

# Add plot details
plt.title('Correlation between Total Cases and Total Deaths (by Country)', fontsize=14)
plt.xlabel('Total Cases')
plt.ylabel('Total Deaths')
plt.xscale('log')
plt.yscale('log')  # Optional: Use log scale for better readability
plt.tight_layout()
plt.show()
```

```python
#Analyze Vaccination Impact on New Case

# Drop rows with missing vaccination or case data
latest_vax = latest_data.dropna(subset=['people_fully_vaccinated_per_hundred', 'new_cases'])

# Scatter plot to visualize vaccination vs. new cases
plt.figure(figsize=(6,4))
sns.scatterplot(data=latest_vax, x='people_fully_vaccinated_per_hundred', y='new_cases', hue='continent',alpha=0.7, palette='Dark2')

plt.title('Impact of Vaccination on New COVID-19 Cases', fontsize=14)
plt.xlabel('People Fully Vaccinated per 100')
plt.ylabel('New Cases')
plt.yscale('log')  # Log scale helps visualize variation
plt.tight_layout()
plt.show()
```

```python
# Daily COVID-19 Trends by Continent

#Filter relevant columns and drop missing continents
continent_trend = covid_data[['date', 'continent', 'new_cases']]
continent_trend = continent_trend.dropna(subset=['continent'])

# Convert 'date' column to datetime type
continent_trend['date'] = pd.to_datetime(continent_trend['date'])

# Group by date and continent, summing new cases
daily_continent = continent_trend.groupby(['date', 'continent'])['new_cases'].sum().reset_index()

# Plot line graph
plt.figure(figsize=(6,4))
sns.lineplot(data=daily_continent, x='date', y='new_cases', hue='continent', palette='tab10')

plt.title('Daily New COVID-19 Cases by Continent', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('New Cases', fontsize=12)
plt.legend(title='Continent')
plt.tight_layout()
plt.show()
```
